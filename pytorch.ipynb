```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
X= torch.arange(-5,5,0.2).view(-1,1)
func=-5*X
Y=func+1.7*torch.rand(X.size())
plt.plot(X.numpy(),Y.numpy(),'b+',label="func")
plt.plot(X.numpy(),func.numpy(),'r',label="func")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid('True',color='grey')
plt.show()
```

```python
def forward(x):
  return w*x
```

```python
# loss fucntion
def criterion(y_pred,y):
  return torch.mean((y_pred-y)**2)
```

```python
w= torch.tensor(-10.0,requires_grad=True)
step_size=0.1
loss_list=[]
iter=20
```

```python
for i in range(iter):
  Y_pred=forward(X)
  loss=criterion(Y_pred, Y)
  loss_list.append(loss.item())

  #lloking into
  loss.backward()
  w.data =w.data-step_size*w.grad.data
  w.grad.data.zero_()
  print('{},\t{},\t{}'.format(i,loss.item(),w.item()))
```

```python
plt.plot(loss_list,'r')
plt.tight_layout()
plt.grid('True',color='grey')
plt.xlabel("Iteration")
plt.ylabel("loss")
plt.show()
```
